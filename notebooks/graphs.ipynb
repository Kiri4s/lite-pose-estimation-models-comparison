```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_ms = [0.058800346064558404, 0.21455205392837523, 0.025373493026780294]
time_fps = [1/i for i in time_ms]
mAP_50_95 = [52, 49.3, 22.7]
memory = [6.3, 88, 2.9] # MB
radius = [i for i in memory]

models = ["YOLOv11n-pose", "lightweight Openpose", "MoveNet MultiPose Lightning"]
colors = ["orange", "red", "green"]

sizes = [i**2 for i in radius]

fig = plt.figure()
for time, metr, size, model, color in zip(time_fps, mAP_50_95, sizes, models, colors):
    plt.scatter(
        time, 
        metr, 
        s=size,  # Marker size (adjust as needed)
        c=color,
        label=model,
        edgecolors='black',
        alpha=0.7
    )

#plt.scatter(time_fps, mAP_50_95, s=sizes, alpha=0.6, c=['yellow', 'red', 'green'], edgecolors='black')
plt.xlabel('latency, fps')
plt.ylabel('$mAP_{50:95}$, %')
plt.grid()
#plt.title('models comparison')
plt.legend(loc='lower left', markerscale=0, labelcolor=colors)
```

```python
fig.savefig("./models_comparison.png")
```
