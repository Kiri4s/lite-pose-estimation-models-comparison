# Plot results

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_ms = [0.058800346064558404, 0.21455205392837523, 0.025373493026780294]
time_fps = [1/i for i in time_ms]
mAP_50_95 = [52, 49.3, 22.7]
memory = [6.3, 88, 2.9] # MB
radius = [i for i in memory]

models = ["YOLOv11n-pose", "Lightweight Openpose", "MoveNet MultiPose Lightning"]
colors = ["orange", "red", "green"]

sizes = [i*10 for i in radius]

fig = plt.figure(figsize=(7,5))
for time, metr, size, model, color in zip(time_fps, mAP_50_95, sizes, models, colors):
    plt.scatter(
        time, 
        metr, 
        s=size,  # Marker size (adjust as needed)
        c=color,
        label=model,
        edgecolors='black',
        alpha=0.7
    )

#plt.scatter(time_fps, mAP_50_95, s=sizes, alpha=0.6, c=['yellow', 'red', 'green'], edgecolors='black')
plt.xlabel('latency, fps')
plt.ylabel('$mAP_{50:95}$, %')
plt.grid()
fig.tight_layout()
#plt.title('models comparison')
#plt.text(0.5, 0.5, 'Рисунок 1. Результаты экспериментов', ha='center', va='center')
plt.figtext(0.5, -0.03, "Рисунок 9. Сравнение моделей на COCO val2017", ha="center", fontsize=12)
plt.legend(loc='lower left', markerscale=0, labelcolor=colors)
```

```python
fig.savefig("../../results/models_comparison.pdf", bbox_inches = "tight")
```

```python
pck = []
acc = []
with open('../../results/yolo.txt', 'r') as f:
    yolo = f.readlines()
pck.append([float(yolo[i]) for i in range(1,18)])
acc.append([float(yolo[i]) for i in range(20,37)])
with open('../../results/lightopenpose.txt', 'r') as f:
    yolo = f.readlines()
pck.append([float(yolo[i]) for i in range(1,18)])
acc.append([float(yolo[i]) for i in range(20,37)])
with open('../../results/movenet.txt', 'r') as f:
    yolo = f.readlines()
pck.append([float(yolo[i]) for i in range(1,18)])
acc.append([float(yolo[i]) for i in range(20,37)])
```

```python
x = np.arange(1, 17+1)
models = ['YOLOv11n-pose', 'Lightweight Openpose', 'Movenet Multipose Lightning']
#keypoint_names = ['нос', 'левый глаз','правый глаз', 'левое ухо', 'правое ухо',
#                  'левое плечо', 'правое плечо', 'левый локоть', 'правый локоть',
#                  'левое запястье', 'правое запястье', 'левое бедро', 'правое бедро', 
#                  'левое колено', 'правое колено', 'левая лодыжка', 'правая лодыжка']
keypoint_names = [
        'nose', 'left_eye', 'right_eye', 'left_ear', 'right_ear',
        'left_shoulder', 'right_shoulder', 'left_elbow', 'right_elbow',
        'left_wrist', 'right_wrist', 'left_hip', 'right_hip',
        'left_knee', 'right_knee', 'left_ankle', 'right_ankle'
    ]
colors = ['tab:orange', 'tab:blue', 'tab:green']

pckarr = np.array(pck)
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')
for m, measurement, c in zip(models,pckarr, colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, np.array(measurement), width, label=m, color=c)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

plt.figtext(0.5, -0.03, "Рисунок 10. Сравнение моделей на COCO val2017 по ключевым точкам", ha="center", fontsize=12)
plt.grid()
ax.set_axisbelow(True)
ax.set_ylabel('PCK@0.1')
ax.set_xticks(x + width, keypoint_names, rotation=-45)
ax.legend(loc='upper right')
ax.set_ylim(0.55, 1)
plt.savefig('../../results/pck.pdf', dpi=300, bbox_inches = "tight")
```

```python
x = np.arange(1, 17+1)
models = ['YOLOv11n-pose', 'Lightweight Openpose', 'Movenet Multipose Lightning']
#keypoint_names = ['нос', 'левый глаз','правый глаз', 'левое ухо', 'правое ухо',
#                  'левое плечо', 'правое плечо', 'левый локоть', 'правый локоть',
#                  'левое запястье', 'правое запястье', 'левое бедро', 'правое бедро', 
#                  'левое колено', 'правое колено', 'левая лодыжка', 'правая лодыжка']
keypoint_names = [
        'nose', 'left_eye', 'right_eye', 'left_ear', 'right_ear',
        'left_shoulder', 'right_shoulder', 'left_elbow', 'right_elbow',
        'left_wrist', 'right_wrist', 'left_hip', 'right_hip',
        'left_knee', 'right_knee', 'left_ankle', 'right_ankle'
    ]
colors = ['tab:orange', 'tab:blue', 'tab:green']

pckarr = np.array(acc)
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')
for m, measurement, c in zip(models,pckarr, colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, np.array(measurement), width, label=m, color=c)
    #ax.bar_label(rects, padding=3)
    multiplier += 1

plt.figtext(0.5, -0.03, "Рисунок 11. Сравнение моделей на COCO val2017 по ключевым точкам", ha="center", fontsize=12)
ax.set_ylabel('Accuracy (10px)')
plt.grid()
ax.set_axisbelow(True)
ax.set_xticks(x + width, keypoint_names, rotation=-45)
ax.legend(loc='upper right')
ax.set_ylim(0.35, 1)
plt.savefig('../../results/acc.pdf', dpi=300, bbox_inches = "tight")
```
